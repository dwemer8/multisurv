# Data info by cancer type

```python
labels = pd.read_csv('../data/labels.tsv', sep='\t')
labels.head()
```

```python
cancer_types = sorted(labels.project_id.unique())
cancer_types += ['All']
columns = ['n', '% censored', 'min_time', 'max_time', 'median_time']

cancer_type_info = {}

for column in columns:
    cancer_type_info[column] = {cancer_type: None for cancer_type in cancer_types}
    
cancer_type_info = pd.DataFrame.from_dict(cancer_type_info)
```

```python
def get_stats(data, in_days=True):
    stats = data.min(), data.max(), data.median()
#     stats = tuple([round(x , 2) for x in stats])
    
    if in_days:
        stats = tuple([x * 365 for x in stats])

    return stats

for cancer_type in sorted(labels.project_id.unique()):
    cancer_labels = labels.loc[labels.project_id == cancer_type]
    cancer_type_info['n'][cancer_type] = len(cancer_labels)

    n_cens = len(cancer_labels.loc[cancer_labels.event == 0])
    percent_cens = (n_cens * 100 / len(cancer_labels))
    cancer_type_info['% censored'][cancer_type] = round(percent_cens, 1)

    (cancer_type_info['min_time'][cancer_type],
     cancer_type_info['max_time'][cancer_type],
     cancer_type_info['median_time'][cancer_type]) = get_stats(cancer_labels['time'], True)
```

```python
# All cancer types
cancer_labels = labels
cancer_type = 'All'

cancer_type_info['n'][cancer_type] = len(cancer_labels)

n_cens = len(cancer_labels.loc[cancer_labels.event == 0])
percent_cens = (n_cens * 100 / len(cancer_labels))
cancer_type_info['% censored'][cancer_type] = round(percent_cens, 1)

(cancer_type_info['min_time'][cancer_type],
 cancer_type_info['max_time'][cancer_type],
 cancer_type_info['median_time'][cancer_type]) = get_stats(cancer_labels['time'], True)
```
